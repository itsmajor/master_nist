```python
from openpyxl import load_workbook
import os

# Zielverzeichnisse
directories = [
    r"output\xlsx\liboqs_scaled",
    r"output\xlsx\nist_scaled"
]

# Öffne jede Datei und ersetze "0" als Text
for dir_path in directories:
    for filename in os.listdir(dir_path):
        if filename.endswith(".xlsx"):
            file_path = os.path.join(dir_path, filename)
            wb = load_workbook(file_path)
            replaced = False

            for sheet in wb.worksheets:
                for row in sheet.iter_rows():
                    for cell in row:
                        if isinstance(cell.value, str) and cell.value.strip() == "0":
                            cell.value = "-"
                            replaced = True

            if replaced:
                print(f"✔ Ersetzt: {file_path}")
                wb.save(file_path)
            else:
                print(f"– Keine '0' gefunden: {file_path}")

print("Fertig – Zellen mit 0 wurden ersetzt.")
```
